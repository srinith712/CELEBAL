**Q1**

 What are the key limitations of traditional MapReduce that make Spark a preferred choice for modern big data processing?

**Ans:**

Traditional MapReduce like Apache Hadoop stores intermediate results in disk after each stage leads to high I/O operations and slow execution.This is inefficient for iterative where read and write operations should perform repeatedly.

Apache spark over comes this by using in-memory computing where data will be in memory during processing.This imprves speed and performance and suitable for big data analytics

**Q2**

 Explain how Spark uses In-Memory Computing to speed up iterative machine learning algorithms compared to disk-based systems.

**Ans**

Spark stores intermediate data in RAM instead of writing it in disk.Machine learning algorithms process the same dataset multiple times so it avoid repeated disk access.these reduces execution time and improves performance compared to disk-based systems 

```python
#importing dataset by creating a volume in databricks 
df=spark.read.csv("/Volumes/assignment5/week5/dataset_volume/dataset.csv",header=True,inferSchema=True)
```

```python
df.display()
```

```python
# Q3:Write a code snippet to remove all duplicate rows from a DataFrame based on a specific set of columns: user_id and transaction_date.

df=df.dropDuplicates(["user_id","transaction_date"])
df.display()
#These remove duplicates based on specific columns (keeps only first occurence)
```

```python
#Q4:Given a DataFrame df_sales, write a query to filter for rows where the region is 'West' and then group by product_category to find the average sale_amount. 
df.filter(df.region=="West").groupBy("product_category").avg("sale_amount").show()
# These Query first filters region West then group the records based on product_category and then finds the average sale_amount
```

```python
#Q5:What is the difference between .na.drop() and .na.fill()? Provide a code example of filling null values in a status column with the string 'Unknown'.
#Ans
# .na.drop() or dropna() removes rows containing null values
# .na.fill() or fillna() replaces null values with a specified value
df=df.fillna("Unknown",subset=["status"])
df.display()
# we can also use df.na.fill() or df.fillna({"status":"Unknown"})
```

```python
#Q6:Write a query to find the total count of records for each city in a DataFrame, but only for cities where the count is greater than 100. 
#Ans
#To use count function we should import it using the below statement
from pyspark.sql.functions import count
df.groupBy("city").agg(count("*")).filter("count(*)>100").show()
#These query groups records based on city then uses aggregate function to count the no of records and then filters the records based on count
```

**Q9**

 When cleaning a dataset, why is it often better to handle null values before performing mathematical aggregations like sum() or avg()?

**Ans**

Null Values can lead to inaccurate calculations or misiing results during aggregation.Handling missing values before performing actions aggreagte functions make calculation meaningful.This improves accuracy.

 

```python
#Q10:Write the code to revise a column named raw_timestamp by casting it to a TimestampType and renaming it to event_time.
from pyspark.sql.functions import col
from pyspark.sql.types import TimestampType
df=df.withColumn(
    "event_time",
    col("raw_timestamp").cast(TimestampType())
).drop("raw_timestamp")
display(df)
```

**Q11**

 Explain the "Shuffle" process that occurs during a grouping operation. Why is it considered a wide transformation?

**Ans**

Shuffle is the process of distributing data across partitions so that records with same grouping key are bought together.Operations like groupBy require this shuffle,making theem wide transforamtions.since shuffle involves network communication and data transfer it is more expensive.Minimizing shuffle operations improves spark performance.

```python
#Q12:Write a code snippet that identifies and removes rows where the email column contains null values OR the username is an empty string.
df.filter((df.email.isNotNull())&(df.username!="")).display()
#These query keeps records with valid email and non empty username
```

```python
#Q13:How do you use the .agg() function to calculate multiple statistics at once, such as the min, max, and mean of the price column?
from pyspark.sql.functions import min,max,avg
df.agg(min("price"),max("price"),avg("price")).show()
#we import those aggregate functions from sql.functions
```

**Q14**

 In the context of cleaning a dataset, what is the risk of using inferSchema=true when your source data contains messy or inconsistent date formats? 

**Ans**

When date formats are inconsistent (inferSchema=True) may incorrectly identify the column as string instead of timestamp.This can cause errors during filtering,sorting or date based calculations.It is often safer to define the schema explicitly or clean the data before loading it.

```python
#Q15:Write a final processing pipeline that: 

#i)Filters out duplicates. 

#ii)Fills null prices with 0. 

#iii)Groups by store_id to calculate total revenue.
from pyspark.sql.functions import sum
df.dropDuplicates().fillna({"price":0}).groupBy("store_id").agg(sum("price")).show()
#This pipeline combines data cleaning  and aggregation to produce the total revenue for each store
```
